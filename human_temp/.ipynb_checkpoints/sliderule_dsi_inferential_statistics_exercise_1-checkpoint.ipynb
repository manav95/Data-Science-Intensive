```python
import pandas as pd
import numpy as np
import math
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set(color_codes=True)

np.random.seed(sum(map(ord, "distributions")))
df = pd.read_csv('data/human_body_temperature.csv')
df
```

```python
# Your work here.
#lets use a histogram to plot the distribution of body temperatures
sns.distplot(df.temperature, hist=True, kde=True)
```

1. It looks normally distributed, perfect for us to apply the Central Limit Theorem and infer about humans themselves.

2. The sample size is 130, so it is greater than 30. We can assume 
independence given that these are samples taken of individual human beings. Thus it is very unlikely the temperatures affect each other.

3. I would use a one-sample test because the sample is of data collected for a group of people from the wider population. Since the sample size is greater than 30, we will use the z test. Here are the results for a z test on the means.

Ho: mean body temp = 98.6 F
Ha: mean body temp != 98.6 F

```python
mean = df.temperature.mean()
std = df.temperature.std()/math.sqrt(130)

print(mean)
print(std)
```

```python
stats.norm.cdf(mean, 98.6, std)
```

It looks like we have observed a significant difference between the sample mean and 98.6. Hence we reject the null hypothesis due to a p-value significantly below .05. Let us repeat this with a t-test, however.

```python
stats.t.cdf(x=mean, df=129, loc=98.6)
```

The t test itself yields a p-value of .363, indicating that it would not yield a significant result. The t distribution is more skewed to the right, making it harder for a value to be considered relatively extreme in the t distribution.

4. 
For this, let us construct a 95% confidence interval for the body temperature in the sample.

```python
interval = stats.norm.interval(.95, loc=mean, scale=stats.sem(df.temperature))
interval
```

With a margin of error of .125 F, it seems that any value outside of this range could be considere
reasonably abnormal. The range is narrow however, and it would seem that our 95% confidence in this 
interval is likely to be mistaken.

5.We will examine this question by dividing the sample into two groups and then using them as representatives of the male and female populations. Each is 65 in length and both groups are independent of each other, so we still satisfy the CLT. We will then use a two-sample, two-sided z test for a difference between the means. The hypotheses are:

Ho: male mean = female mean
Ha: male mean != female mean

```python
males = df[df.gender == 'M']
females = df[df.gender == 'F']


maleTempMean = males.temperature.mean()
femaleTempMean = females.temperature.mean()

print (males.shape)
print (females.shape)
print(maleTempMean)
print(femaleTempMean)
print (males.temperature.std()/math.sqrt(65))
print (females.temperature.std()/math.sqrt(65))
```

```python
import statsmodels.stats.weightstats as ws

ws.ztest(males.temperature.as_matrix(), females.temperature.as_matrix(), alternative='two-sided', usevar='pooled')
```

With a p-value of .02229, we are able to conclude that there is a significant difference between males and females with regards to body temperature. Women tend to have a higher body temperature than men do, at around 98.4. This difference we observed falls under the alpha of .05, so we are 95% confident this is not due to chance and reject the null hypothesis of no difference.
