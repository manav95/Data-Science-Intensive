### Make a bar plot of the months in which movies with "Christmas" in their title tend to be released in the USA.

```python
usChristmasRelease = release_dates[(release_dates.country == 'USA') & (release_dates.title.str.contains("Christmas"))]
usChristmasRelease.groupby([usChristmasRelease.date.dt.month]).size().plot(kind='bar')
```

### Make a bar plot of the months in which movies whose titles start with "The Hobbit" are released in the USA.

```python
usChristmasRelease = release_dates[(release_dates.country == 'USA') & (release_dates.title.str.startswith("The Hobbit"))]
usChristmasRelease.groupby([usChristmasRelease.date.dt.month]).size().plot(kind='bar')
```

### Make a bar plot of the day of the week on which movies with "Romance" in their title tend to be released in the USA.

```python
usChristmasRelease = release_dates[(release_dates.country == 'USA') & (release_dates.title.str.contains("Romance"))]
usChristmasRelease.groupby([usChristmasRelease.date.dt.weekday_name]).size().plot(kind='bar')
```

### Make a bar plot of the day of the week on which movies with "Action" in their title tend to be released in the USA.

```python
usChristmasRelease = release_dates[(release_dates.country == 'USA') & (release_dates.title.str.contains("Action"))]
usChristmasRelease.groupby([usChristmasRelease.date.dt.weekday_name]).size().plot(kind='bar')
```
