# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
%matplotlib inline

import sys
print(sys.version)
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import bokeh.plotting as bkp
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# read in readmissions data provided
hospital_read_df = pd.read_csv('data/cms_hospital_readmissions.csv')
hospital_read_df.head()
```

****
## Preliminary Analysis

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
clean_hospital_read_df.loc[:, 'Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)
clean_hospital_read_df = clean_hospital_read_df.sort_values('Number of Discharges')
```

```python
# generate a scatterplot for number of discharges vs. excess rate of readmissions
# lists work better with matplotlib scatterplot function
x = [a for a in clean_hospital_read_df['Number of Discharges'][81:-3]]
y = list(clean_hospital_read_df['Excess Readmission Ratio'][81:-3])

fig, ax = plt.subplots(figsize=(8,5))
slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x,y)

ax.scatter(x, y,alpha=0.2)

ax.fill_between([0,350], 1.15, 2, facecolor='red', alpha = .15, interpolate=True)
ax.fill_between([800,2500], .5, .95, facecolor='green', alpha = .15, interpolate=True)

ax.set_xlim([0, max(x)])
ax.set_xlabel('Number of discharges', fontsize=12)
ax.set_ylabel('Excess rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=14)

ax.grid(True)
fig.tight_layout()
print(r_value ** 2)
```

There is no correlation to be found here. Furthermore, there are outliers in the red and green areas.

The correlations seem a little better for some of the categories, but overall not promising. It seems that capacity has little correlation with excess readmissions rate. There does seem to be a difference between the categories of the conditions and the number of instances in the red area and green area. We will do some testing on the hip/knee and see if there is a significant difference in the mean for the ones below 200 and others above 800.

```python
belowTwoHundred = hipKnee[hipKnee['Number of Discharges'] <= 200]
aboveEightHundred = hipKnee[hipKnee['Number of Discharges'] >= 800]
```

```python
average = hipKnee['Excess Readmission Ratio'].mean()
print(average)
print(belowTwoHundred['Excess Readmission Ratio'].mean())
print(aboveEightHundred['Excess Readmission Ratio'].mean())
```

Lets run two t tests to see if there is significance between the two groups. There is a substantial practical difference that we can observe. Our null hypotheses are of no difference in the mean excess readmissions ratio.

```python
firstTest, firstPVal = scipy.stats.ttest_ind(hipKnee['Excess Readmission Ratio'], belowTwoHundred['Excess Readmission Ratio'])
secTest, secPVal = scipy.stats.ttest_ind(hipKnee['Excess Readmission Ratio'], aboveEightHundred['Excess Readmission Ratio'])
print(firstPVal)
print(secPVal)
```

Hence we can reject the null hypotheses in both cases, with a significance level of .01, and say there is a significant difference in both cases.
